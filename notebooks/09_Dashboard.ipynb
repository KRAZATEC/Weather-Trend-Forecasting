# 09 - Dashboard (Preview)
The full interactive dashboard lives in `dashboard/app.py` and is built with Streamlit. This notebook renders a static preview of the same views using Plotly, so the key visuals are inspectable without running the Streamlit server.

````python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
import pandas as pd
import plotly.express as px
from src import config

df = pd.read_csv(config.FEATURED_DATA_FILE, parse_dates=[config.DATETIME_COL])
latest = df.sort_values(config.DATETIME_COL).groupby('location_name').tail(1)
fig = px.scatter_geo(latest, lat='latitude', lon='longitude', color='temperature_celsius',
                     hover_name='location_name', color_continuous_scale='RdYlBu_r',
                     projection='natural earth', title='Global Temperature Snapshot')
fig.show()
````

````python
fig2 = px.line(df.set_index(config.DATETIME_COL)['temperature_celsius'].resample('D').mean(),
               title='Global Daily Average Temperature Trend')
fig2.show()
````
